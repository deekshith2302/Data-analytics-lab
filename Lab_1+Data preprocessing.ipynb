Imputation

Imputation is the process of replacing missing data with substituted values.

In data science and statistics, datasets often have gapsâ€”cells where information was not recorded or was lost. Because most machine learning algorithms cannot process blank values (NaN/Null), you must handle them. Instead of deleting the entire row or column (which loses valuable information), imputation fills these gaps with estimated values.

The goal of imputation is not just to "fill the hole," but to do so in a way that preserves the overall statistical relationships (like the mean, variance, and correlations) of the dataset.

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_excel("Employee_data.xlsx")
```

```python
print("Original DataFrame with Missing Values:")
print(df)
```

```python
print(f"Missing values per column:\n{df.isnull().sum()}\n")
```

```python
df_imputed_combined = df.copy()

median_age = df_imputed_combined['Age'].median()
df_imputed_combined['Age'] = df_imputed_combined['Age'].fillna(median_age)

median_salary = df_imputed_combined['Salary'].median()
df_imputed_combined['Salary'] = df_imputed_combined['Salary'].fillna(median_salary)

print("DataFrame after imputing both missing Age and Salary with their respective medians:")
print(df_imputed_combined)
print("\n")
```

```python
Q1 = df_imputed_combined['Age'].quantile(0.25)
Q3 = df_imputed_combined['Age'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print(f"Age Bounds: {lower_bound} to {upper_bound}")

# Filter the data
df_outliers_age_removed = df_imputed_combined[
    (df_imputed_combined['Age'] >= lower_bound) &
    (df_imputed_combined['Age'] <= upper_bound)
]

print("Rows removed (Noise) from Age column:")
print(df_imputed_combined[~df_imputed_combined.index.isin(df_outliers_age_removed.index)])
print("\n")
print("DataFrame after Age noise removal:")
df_imputed_combined = df_outliers_age_removed.copy() # Update df_imputed_combined
print(df_imputed_combined)
```

```python
Q1 = df_imputed['Salary'].quantile(0.25)
Q3 = df_imputed['Salary'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print(f"Age Bounds: {lower_bound} to {upper_bound}")

# Filter the data
df_clean_noise = df_imputed[
    (df_imputed['Salary'] >= lower_bound) &
    (df_imputed['Salary'] <= upper_bound)
]

print("Rows removed (Noise):")
print(df_imputed[~df_imputed.index.isin(df_clean_noise.index)])
print("\n")
print("Dataset after noise removal")
print(df_clean_noise)
```

```python
duplicates = df_clean_noise[df_clean_noise.duplicated(keep=False)]
print("Duplicate Rows found:")
print(duplicates)

# Remove duplicates (keep the first occurrence)
df_final = df_clean_noise.drop_duplicates(keep='first')

print("\n--- FINAL CLEANED DATAFRAME ---")
print(df_final)
```
